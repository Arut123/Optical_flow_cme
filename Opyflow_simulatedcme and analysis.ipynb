```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib 
df1=pd.read_csv('good_features_positions_and_opt_flow_from_frame0to39_with_step_1_and_shift_2.csv')
df1.head()
```

```python
df2=df1.drop(['Time_A','Time_B','X','Y'],axis='columns')
df2.head()
```

```python
df3=df2.dropna()
df3.isnull().sum()
df3.shape
```

```python
df4=df3[~(df3.Vx==0)]
df4.shape
```

```python
df4['Vx']=df4['Vx'].abs()
df4['Vy']=df4['Vy'].abs()
df4.head()
```

```python
df4['Vx2']=df4['Vx']**2
df4['Vx2']
```

```python
df4['Vy2']=df4['Vy']**2
df4['Vy2']
```

```python
df4['norm']=(df4['Vx2']+df4['Vy2'])**0.5
df4['norm']
```
